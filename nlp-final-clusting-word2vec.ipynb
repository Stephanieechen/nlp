```python
import nltk
import pandas as pd
import numpy as np
import string
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')
df = pd.read_csv('nlp.csv')
df1 = df[df['overall'] > 3]
df1.reset_index(inplace = True)
df1.drop(['index','Unnamed: 0'],axis =1, inplace =True)
```

```python
df1.dropna(subset = ['reviewText'],inplace = True)
```

```python
a = list()
b = df1['reviewText'].values.tolist()
for i in b:
    token = RegexpTokenizer(r'\w+')
    tokens = token.tokenize(i)
    words = [w.lower() for w in tokens]
    stopwords = nltk.corpus.stopwords.words('english')
    newStopWords = ["however",'unless','sometimes','usually','finally','definitely','stylus']
    stopwords.extend(newStopWords)
    words =[w for w in words if not w in stopwords]
    #words = [snowball_stemmer.stem(w) for w in words]
    tags = nltk.pos_tag(words)
    adj = [word for word,pos in tags if (pos == 'RB' or pos == 'JJ')]
    a.append(adj)
```

```python
from gensim.models.word2vec import Word2Vec
model = Word2Vec(a,min_count=5)
word_vectors = model.wv
```

```python
word_vectors.most_similar(positive = ['battery','good'],topn = 10)
```
